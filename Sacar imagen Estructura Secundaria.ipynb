```python
#Las librerias clasicas
import numpy as np
import pandas as pd

import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio
```

```python
x = pd.read_csv("assets/ssecondary_sus2.dat",delimiter="\s+",index_col=0)
# x = pd.read_csv("assets/ssecondary_bms_comp.dat",delimiter="\s+",index_col=0)

df  = pd.read_csv('assets/numerado.csv')
# aa = df[df["Sistema"]=="Estructura Nativa"]["Subunidad"].values+": "+df[df["Sistema"]=="Estructura Nativa"]["Aminoacido"].values+"-"+df[df["Sistema"]=="Estructura Nativa"]["Corregida"].values.astype("str")
aa = df[df["Sistema"]=="Complejo Enzima-Sustrato"]["Subunidad"].values+": "+df[df["Sistema"]=="Complejo Enzima-Sustrato"]["Aminoacido"].values+"-"+df[df["Sistema"]=="Complejo Enzima-Sustrato"]["Corregida"].values.astype("str")
# aa[760] = "PSEN: PRO-166"
```

```python
fig = go.Figure(data=go.Heatmap(z=x.values[:,1:].T,x=x.index.values/100,y=aa[:-1],
    colorscale = ["rgba(229,236,246,255)","steelblue","midnightblue","maroon","firebrick","darkgoldenrod","palegoldenrod", "beige"],
    colorbar = dict(thickness=10, tickvals=[0,1,2,3,4,5,6,7], ticktext=["None",'Para',"Anti","3-10","Alpha","Pi","Turn", "Bend"])))

# fig.update_layout(title='Complejo Enzima Sustrato E-2012',title_x=0.5,title_y=0.99,autosize=True,width=1000,height=4000,yaxis = dict(tickfont = dict(size=10)))
fig.update_layout(title='Estructura secundaria Î³-secretasa Sus Replica 2',title_x=0.5,title_y=0.99,autosize=True,yaxis = dict(tickfont = dict(size=10)))

fig.update_traces(colorbar_orientation='h', selector=dict(type='heatmap'))
fig.update_xaxes(title="Tiempo (ns)")
fig.update_yaxes(title="Residuos")
fig.show()
```

```python
pio.write_image(fig, "figure/Estructura secundaria_sus_replica2.svg", width=3*300, height=12*300, scale=1)
# pio.write_image(fig, "../TM6mut.png", width=1.5*300, height=2*300, scale=1)
# pio.write_image(fig, "../TM6apo.png", width=1.5*300, height=2*300, scale=1)
```

```python

ss_ss = pd.read_csv('assets/ssecondary_mut.dat.sum',sep="\s+")
imagen = "Estructura secundaria_mut.svg"
title="Estructura Secundaria Estructura Nativa L166P"
amino =aa

#Este chunk de codigo es para dejar solo la estructura mas representativa en las laminas beta
ss_ss["Fracc" ]= ss_ss["Para"].values+ss_ss["Anti"].values+ss_ss["3-10"].values+ss_ss["Alpha"].values+ss_ss["Pi"].values+ss_ss["Turn"].values+ss_ss["Bend"].values
mn = ss_ss[ss_ss["Fracc"]>1.5].iloc[:,1:3].values.argmin(axis=1) +1
ind = ss_ss[ss_ss["Fracc"]>1.5].index
for i in range(len(mn)):
    ss_ss.iloc[ind[i],mn[i]] = 0
    ss_ss["Fracc" ]= ss_ss["Para"].values+ss_ss["Anti"].values+ss_ss["3-10"].values+ss_ss["Alpha"].values+ss_ss["Pi"].values+ss_ss["Turn"].values+ss_ss["Bend"].values
#aca acaba el chunk

colores=["steelblue","midnightblue","maroon","firebrick","darkgoldenrod","palegoldenrod", "beige"]
tipos =['Para',"Anti","3-10","Alpha","Pi","Turn", "Bend"]
fig = go.Figure()
for i in range(len(tipos)):
    fig.add_traces(go.Bar(x=amino, y=ss_ss[tipos[i]],name=tipos[i],
                     marker_color=colores[i]))
fig.update_layout(title_x=0.5,title_text=title,barmode="stack",width=2000,height=400,)
fig.update_xaxes(title="Residuos")
fig.update_yaxes(title="Probabilidad")
```
